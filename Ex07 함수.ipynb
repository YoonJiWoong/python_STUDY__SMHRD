```python
num1=int(input("첫 번째 정수 입력 >>"))
num2=int(input("첫 번째 정수 입력 >>"))
op=input("연산자 입력(+,-)>>")
result = cal(num1,num2,op)
print(f"결과 : {result}")
```

```python
# 매개변수의 기본값 설정하기
def power_of_N(num, power=2):
    return num**power
    
```
