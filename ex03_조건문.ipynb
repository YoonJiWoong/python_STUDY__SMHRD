#### 조건문
- 상황에 따라 다르게 처리하는 구문
- 종류 : if, elif, else


```python
if True :
    print('실행문장1')
print('실행문장2')

if False :
    print('실행문장1')
    print('실행문장2')
print('실행문장3')

```

```python
money = 9000
if  money>=10000:
    print("택시를 탄다")
else:
    print("택시를 안탄다")
  
   
```

```python
num1 = int(input('정수를 입력>> : '))
if 100>=num1>=90:
    print(f'{num1}점은 A학점 입니다.')
elif 90>num1>=80:
    print(f'{num1}점은 B학점 입니다.')
elif 80>num1>=70:
    print(f'{num1}점은 C학점 입니다.')
elif 70>num1>=60:
    print(f'{num1}점은 D학점 입니다.')
elif 60>num1:
    print(f'{num1}점은 F학점 입니다.')
    
```

```python
score = int(input('정수를 입력>> : '))
if 100>=score>=90:
    grade='A'
elif 90>score>=80:
    grade='B'
elif 80>score>=70:
    grade='C'
elif 70>score>=60:
    grade='D'
elif 60>score:
    grade='F'
print(f'{score}점은 {grade}학점 입니다.')
```
