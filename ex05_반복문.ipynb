```python
score_list = [90,45,70,60,55]
j=0
for i in score_list:
    
    if i>=60:
        print(f'{j+1}  .')
    else:
        print(f'{j+1}  .')
    j+=1
        
   
```

```python

score_list = [90,45,70,60,55]

for i in score_list:
    
    if i>=60:
        print(f'{score_list.index(i)+1}  .')
    else:
        print(f'{score_list.index(i)+1}  .')

```
