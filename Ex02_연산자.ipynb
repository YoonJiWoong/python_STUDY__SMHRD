#### 연산자
- 종류 : 산술 연산자, 지수 연산자, 대입 연산자, 삼항 연산자, 비교 연산자, 논리 연산자
- 나누기 연산자 :  /(나누기),  //(몫),  %(나머지)

```python
num1 = 10
num2 = 7
print(num1/num2) # 나누기
print(num1//num2) # 나누기 몫
print(num1%num2) # 나눴을 때 나머지
```

```python
#문자열 더하기
str1 = '안녕'
str2 = '하세요'
print(str1,str2)
print(str1+str2)
```

```python
str3 = '10'
str4 = '7'
print(str3,str4)
print(str3+str4)
print(10+7)
```

```python
# 숫자 + 문자
num1 = 10
str2 = "7"
# 서로 다른 데이터 타입의 연산을 지원하지 않는다.
print(num1 + str2)
# -> 데이터 타입을 맞춰 주자!
```

```python
#숫자 + 숫자
print(num1 + int(str2))
#문자 + 문자
print(str(num1)+str2)
```

```python
str2 = int(str2)
type(str2)
```

#### 키보드로부터 입력받기 : input
- 입력한 값을 문자열로 돌려준다! (반환)

```python
num = input('정수를 입력해주세요 :')
num #문자열
```

```python
# 문자열로 반환 -> 산술연산X
# 해결책 : 문자열을 정수로 반환
num1 = int(input('정수를 입력해주세요 : '))
num1
```

```python
type(num1)
```

```python
num1 = int(input('정수를 입력해주세요 : '))
num2 = int(input('정수를 입력해주세요 : '))
print('더하기 결과 :', (num1+num2))
print('빼기 결과 :'+str((num1-num2)))
print('곱하기 결과 :',(num1*num2))
print('나누기 결과 :',(num1/num2))
```

```python
# 지수 연산자 : 밑수**지수
# 2의 3승은? 8
print(2**3)
print(4**4)
```

```python
# 복합 대입 연산자
num3 = 27
num3 +=3 # num3 = num3 + 3
num3 +=3 # num3 = num3 + 3
num3 +=3 # num3 = num3 + 3
print(num3)
```

```python
# 치환
a=3
b=20
print(a,b)

#기존 치환 방법
temp=a
a=b
b=temp
print(a,b)

# 파이썬 치환
a,b=b,a
print(a,b)
```

```python
# 비교연산자
# 반환값은 논리값(True/False)
c = 4
d = 15
print(c>b)
print(c<b)
print(c ==b)
print(c !=b)
```

```python
# not 연산자 - 논리값을 뒤집어 준다.
print(not True)
print(not 3 >=20)
```

```python
# and 연산자 - 모두 True 여야 True
print(True and True)
print(True and False)
print(3 <20 and 3 >=20)
print(3 <20 and 3 !=20)
print(3 >=20 and 3 ==20)
```

```python
# or 연산자 - 하나라도 True 이면 True
print(True and True)
print(True and False)
print(3 <20 or 3 >=20)
print(3 <20 or 3 !=20)
print(3 >=20 or 3 ==20)
```

```python
#삼항 연산자
# a if 조건식 else b
score = 70
print('합격'if score >= 60 else '불합격' )
```

```python
num = int(input('정수를 입력>> '))
result = '짝수' if num%2==0 else '홀수'
print(f'{num}은(는) {result}입니다')
```
