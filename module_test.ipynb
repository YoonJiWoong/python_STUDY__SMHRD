```python
def number_sum(n1,n2):
    result=n1+n2
    return result
```

```python
def cal(num1,num2,op):
    """덧셈과 뺄셈을 계산하는 함수"""
    if op=="+":
        return num1+num2
    elif op=="-":
        return num1-num2
```

```python
def add_sub(num1,num2):
    return num1+num2, num1-num2
```
