```python
# 리스트 인덱싱
list1 = [2,5,7,9,10]
print(list1[2])

#temp 변수에 ['a','b','c']
list2 = [1,2,3,['a','b','c']]
temp= list2[3]
print(temp)

#list2에서 'b' 출력하기
print(list2[3][2])   #인덱스 번호는 0부터 이 경우 3이라고 안치게 조심하기!
```

```python
#리스트 슬라이싱
list3 = [4,7,2,1,9]
print(list3[1:3])
print(list3[:3])
print(list3[2:])
print(list3[:1])
print(list3[0]) #결과값의 데이터 타입의 차이 : 인덱싱은 해당값 슬라이싱은 리스트 / 리턴되는 값으 형태다름
```

```python
#리스트 요소 수정
list8=[1,5,7,6,3]
list8[2]=9
print(list8)
list8[1:3]= [7]
print(list8)
```

```python
array=[1,2,3,4,5]

# 5를 7로 변경
array[-1]=7 # array[4]=7
print(array)

# 2를 ['a','b','c']
array[1]=['a','b','c']
print(array)

# 4,7 = d,e,f,g
array[-2:]=['d','e','f','g']
print(array)

# 1을 'h','y','j'
array[:1]=['h','y','j']
print(array)
```

```python
# 리스트 요소 삭제
# - 인덱스를 이용해서 삭제 -> del 키워드를 사용
# - 어떤 값을 삭제할지 명시 -> remove() 함수 사용

list9 = ['a','b','c','d','a']
del list9[2] # 인덱스를 이용해서 삭제
print(list9) 
list9.remove('a') #어떤 값을 삭제할시 명시, a가 2개이상이면 맨 앞에 것만 삭제한다!
print(list9)
```

```python
list9*3
print(list9)
```

```python
lst = ['딸기','바나나','수박','체리','포도']
a =input('검색할 문자를 입력하세요>>')
if a in lst:
    print(f'{a}는 리스트에{lst.index(a)+1}번째 들어있습니다.')
else:
    print(f'{a}는 리스트에 없습니다')
```
