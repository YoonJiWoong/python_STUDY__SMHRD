# 인덱싱과 슬라이싱


```python
s = 'My name is JW'
print(s)
```

```python
print(s[3:7]) #3이상 7미만! 뒤에 오는 인덱스보다 1작은 인덱스 까지 가지고 온다
print(s[-1])
print(s[0:7]) # 0부터 시작할때는 0 생략 가능
print(s[:7])
print(s[8:13])
print(s[8:])
print(s[-4:])
print(s[:])
```

```python
day="2021년 11월 08일의 날씨는 맑음 입니다."
print("날짜 : "+day[:13])
print('날씨 : '+day[-7:-5])
```

```python
#문자열 포매팅 - 문자열 안의 특정한 값을 바꿔야 할 경우가 있을때 사용
sensorValue=12
tmp = '오늘의 온도는 %d도 입니다.' %sensorValue
print(tmp)
```

```python
m = 11
d = 8
tmp = 12.2
str2 = '오늘의 날짜는 %d월 %d일이고, 온도는 %0.2f도 입니다 15%%입니다' %(m,d,tmp) #책 p52.참고/ 자리수는 0.n으로
print(str2)
```

```python
# format 함수를 사용한 포매팅
# 문자열 .format()
# 특정 데이터가 들어가야 하는 자리에 {}를 넣어준다.

m=11
d=8
str3 = '오늘은 {}월 {}일 입니다.'.format(m,d)
print(str3)
```

```python
month =7
day =7
s= "오늘은{}월{}일 입니다".format(month,day) #바로 붙여서 쓴다 .(점)찍고 써야한다!
print(s)
```

```python
# f문자열을 사용한 포매팅
# 문자열 가장 앞에 f 키워드를 붙여준다.
# 특정 데이터가 들어가야할 자리에 중괄호를 넣어주고 
# 중괄호 안에 그 데이터도 같이 담아준다.

m = 11
d = 8
str4 = f'오늘은 {m}월 {d}일 입니다.'
print(str4)
```

#### 문자열 함수

```python
msg3 = "Python is easy and simple."
print(msg3)
# 문자열의 길이 : len(데이터 or 변수)
print(len(msg3)) #length
```

```python
# 양쪽 공백 제거 : strip()
msg4 = '     Python is easy and simple.  '
print(msg4)
print(msg4.strip())
```

```python
msg4 = msg4.strip()
msg4
```

```python
#특정 문자를 다른 문자로 대체 : replace
#변수. replace(이전 문자, 대체할 문자)
print(msg4)
print(msg4.replace('Python','Java'))
# ---> 실제 변수에 저장된 데이터에 영향을 주지 않음.
print(msg4)
# ---> 대체한 결과를 변수에 대입시키면 된다.
msg1 = msg4.replace('Python','Java')
print(msg1)
```
