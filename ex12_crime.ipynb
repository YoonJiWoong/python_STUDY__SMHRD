```python
cr2015 = pd.read_csv('2015.csv',encoding = 'euc-kr',index_col='')
cr2016 = pd.read_csv('2016.csv',encoding = 'euc-kr',index_col='')
cr2017 = pd.read_csv('2017.csv',encoding = 'euc-kr',index_col='')
```

```python
a2015=a.loc[:,'':''].sum(axis=1)
a2015 # 2015
```

```python
a2016=b.loc[:,'':''].sum(axis=1)
a2016 # 2016
```

```python
a2017=c.loc[:,'':''].sum(axis=1)
a2017 # 2017
```

```python
a16_15=(a2016-a2015)/a2015*100
a16_15
```

```python
a17_16=(a2017-a2016)/a2016*100
a17_16
```
