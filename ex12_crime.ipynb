```python
import pandas as pd
```

```python
cr2015 = pd.read_csv('2015.csv',encoding = 'euc-kr',index_col='관서명')
cr2016 = pd.read_csv('2016.csv',encoding = 'euc-kr',index_col='관서명')
cr2017 = pd.read_csv('2017.csv',encoding = 'euc-kr',index_col='관서명')
```

```python
a=cr2015[cr2015['구분']=='발생건수']
a
```

```python
b=cr2016[cr2016['구분']=='발생건수']
b
```

```python
c=cr2017[cr2017['구분']=='발생건수']
c
```

```python
c=c.drop('광주지방경찰청') #행 삭제
c
# c=c.drop('광주지방경찰청', inplace=True) #행 삭제
```

```python
a2015=a.loc[:,'살인':'폭력'].sum(axis=1)
a2015 # 2015총계
```

```python
a2016=b.loc[:,'살인':'폭력'].sum(axis=1)
a2016 # 2016총계
```

```python
a2017=c.loc[:,'살인':'폭력'].sum(axis=1)
a2017 # 2017총계
```

```python
a16_15=(a2016-a2015)/a2015*100
a16_15
```

```python
a17_16=(a2017-a2016)/a2016*100
a17_16
```

```python
s5=pd.concat([a2015,a16_15,a2016,a17_16,a2017], axis=1)
s5.columns=['2015년총계','2015-2016증감율','2016년총계','2017-2016증감율','2017년총계']
s5
```
