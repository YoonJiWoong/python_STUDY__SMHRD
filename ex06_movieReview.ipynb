```python
# for review in data:
#     print(review[1])
n_cnt=p_cnt=0

for review,np in data:
    
    if np == 1:
        p_cnt+=1
    else :
        n_cnt+=1
        
print(f'긍정은 {p_cnt}개')
print(f'부정은 {n_cnt}개')
```

```python
cnt=0
for review,np in data:   
    if "ㅋ" in review:
        cnt+=1
print(f'ㅋ가 포함된 리뷰는 {cnt}개')
```

```python
cnt=0
review_len=0
for review,np in data:
    
    review_len=review_len+len(review)
    

print(f'리뷰의 평균 길이는 {review_len/len(data)} 입니다')
```

```python
review_max=0
review_in=""
for review,np in data:
    if review_max<len(review):
        review_max=len(review)
        review_in=review
        
print(f'리뷰의 최대 길이는{review_max}입니다')
print(review_in)

    
```
