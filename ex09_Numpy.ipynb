```python
data = np.loadtxt("ratings.txt",delimiter="::",dtype="int")
data
```

```python
data[:,0]==1
print(data[:,0]==1)
```

```python
# 1   

user1=data[data[:,0]==1, :]
print(data[data[:,0]==1, :])
#  data[data[:,0]==1] 
```

```python
# 1    

np.mean(user1[:,2])


#arr1 = data[data[:,0]==1, :]
#np.mean(arr1, axis=0, dtype='int')
```

```python
#     

user_list = []

# data[:,0]     
#     
user_uni=np.unique(data[:,0])
 
for i in user_uni:
    #    
    user1=data[data[:,0]==i, :]
    #       
    user_mean=np.mean(user1[:,2]) #    
    user_list.append([i,user_mean])
```

```python
user_list
```

```python
user_meanlist = np.array(user_list)
user_meanlist
```

```python
#     4   
user_meanlist[user_meanlist[:,1]>=4]
#   ...
user_meanlist[user_meanlist[:,1]>=4][:,0].astype(int) #   
```
