## 배열 연산 함수

```python
# 배열 만들기
# 2차원 배열 
# 1 ~ 10까지 수를 담고
# (2,5)의 크기로 배열 만들기

arr3 = np.arange(1,11)
arr3 = arr3.reshape(2,5)
arr3
```

```python
# sum
# 가지고 있는 모든 요소를 계산
np.sum(arr3)
```

```python
# axis = 1 행단위
# axis = 0 열단위
print(np.sum(arr3 ,axis =0))
print(np.sum(arr3 ,axis =1))
```

```python
# mean
# 전체 요소에 대한 평균
np.mean(arr3)
```

```python
# 열 단위의 평균
np.mean(arr3, axis=0)
```

```python
# 행 단위의 평균
np.mean(arr3, axis=1)
```

```python
data = np.loadtxt("ratings.txt",delimiter="::",dtype="int")
data
```

```python
data[:,0]==1
print(data[:,0]==1)
```

```python
# 1번 사용자의 데이터만 가져오기

user1=data[data[:,0]==1, :]
print(data[data[:,0]==1, :])
#  data[data[:,0]==1]도 가능
```

```python
# 1번 사용자의 평점 평균 구하기

np.mean(user1[:,2])


#arr1 = data[data[:,0]==1, :]
#np.mean(arr1, axis=0, dtype='int')
```

```python
# 각 사용자별 평점 평균 구하기

user_list = []

# data[:,0] 데이터의 유저 번호를 다 가져오고
# 중복된 값을 제거한 후 출력
user_uni=np.unique(data[:,0])
 
for i in user_uni:
    # 각 사용자의 정보 저장
    user1=data[data[:,0]==i, :]
    # 저장된 정보에서 평점 값만 가져와서 평균 구하기
    user_mean=np.mean(user1[:,2]) #다른 유저 평점가져와 평점 계산
    user_list.append([i,user_mean])
```

```python
user_list
```

```python
user_meanlist = np.array(user_list)
user_meanlist
```

```python
# 각 사용자별 평점 평균이 4점 이상인 사용자 구하기
user_meanlist[user_meanlist[:,1]>=4]
# 사용자만 구하면 되니깐...
user_meanlist[user_meanlist[:,1]>=4][:,0].astype(int) # 아이디 값만 가져와줘
```
